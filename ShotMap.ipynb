```python
#imports
import pandas as pd
import matplotlib.pyplot as plt
from highlight_text import fig_text
import matplotlib as mpl
from mplsoccer import Pitch, VerticalPitch
```

```python
#read csv file
data = pd.read_csv('shotMapArsenal23042021.csv')
```

```python
data
```

```python
#Get total xg dinamically
xgTotalFloat = 0
#add all xg per shot
for i in range(len(data)):
    xgTotalFloat += data.xg[i]
    #print(data.xg[i])
```

```python
xgTotalFloat
```

```python
def truncate(num, n):
    integer = int(num * (10**n))/(10**n)
    return float(integer)

#Truncate to just 3 decimals
xGTotal = truncate(xgTotalFloat, 3)
xGTotal
```

```python
pitch = VerticalPitch(half=True, pitch_color='#146041', line_color='white',stripe=False)
# pitch = VerticalPitch(half=True, pitch_color='#aabb97', line_color='white',stripe_color='#c2d59d', stripe=True)
#005244
#146041
fig, ax = pitch.draw(figsize=(12,8)) #Cambiar tamaño
pitch.draw(ax=ax)
#I invert the axis to make it so I am viewing it how I want
plt.gca().invert_yaxis()

total_shots = len(data)
fig_text(s=f'Total Shots: {total_shots}', x=.27, y =.75, fontsize=18,color='w')
fig_text(s=f'xG: {xGTotal}',x=.49, y =.75, fontsize=18,color='w')
fig_text(s=f'Goals: 0', x=.68, y =.75, fontsize=18,color='w')

#plot the points, you can use a for loop to plot the different outcomes if you want
plt.scatter(data['y'],data['x'], s=(data.xg*1000),c='#FFE700',alpha=.6)

fig.text(0.68,.045,f'@sportsdatalab / twitter',fontsize=14,color='w')
```
